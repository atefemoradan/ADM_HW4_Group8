```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
soup=[]
for i in range(600):
    url = 'https://www.immobiliare.it/vendita-case/roma/?criterio=rilevanza&pag='+str(i+1)
    soup.append(BeautifulSoup(requests.get(url).text, "html.parser"))
    time.sleep(2)
```

```python
soup[-1].findAll('ul',{'class':'listing-features list-piped'})
```

```python
l1=['piano','bagni','superficie','locali']
df=[]
for x in soup:
    x=x.findAll('ul',{'class':'listing-features list-piped'})
    y=x.findAll('p',{'titolo text-primary'})
    for house in x:
        d={}
        house=[text for text in house.stripped_strings]
        house.reverse()
        for i in l1:
            if i in house:
                index=house.index(i)
                if i=='superficie':
                    d[i]=house[index+3]
                else:
                    d[i]=house[index+1]
                if house[-1].find('€')!=-1:
                    d['prezzo(€)']=house[-1]
                elif house[-2].find('€')!=-1:
                     d['prezzo(€)']=house[-2]
        df.append(d)        -----
```

```python
df1=pd.DataFrame.from_dict(df)
```
